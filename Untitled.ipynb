```python
interaction_matrix = df.pivot_table(
    index='customer_id',
    columns='product_id',
    values='quantity',   # or 'sales_amount'
    aggfunc='sum',
    fill_value=0
)

print(interaction_matrix.head())
```

```python
from sklearn.metrics.pairwise import cosine_similarity
user_similarity = cosine_similarity(interaction_matrix)
similarity_df = pd.DataFrame(user_similarity, 
                             index=interaction_matrix.index, 
                             columns=interaction_matrix.index)
```

```python
similar_customers = similarity_df['C000001'].sort_values(ascending=False).head(6)
print(similar_customers)
```

```python
def recommend_products(target_customer, interaction_matrix, similarity_df, top_n=5):
    # find similar customers (excluding the target itself)
    neighbors = similarity_df[target_customer].sort_values(ascending=False).index[1:6]
    
    #  products bought by neighbors
    neighbor_purchases = interaction_matrix.loc[neighbors].sum()
    
    # exclude products already bought by target
    already_bought = interaction_matrix.loc[target_customer]
    recommendations = neighbor_purchases[already_bought == 0].sort_values(ascending=False).head(top_n)
    
    return recommendations

# Example usage
recommendations = recommend_products('C000001', interaction_matrix, similarity_df, top_n=5)
print("Top Recommendations for C000001:")
print(recommendations)
```

```python
# Join with product names
recommended_products = recommendations.index
product_info = df[df['product_id'].isin(recommended_products)][['product_id','product_name']].drop_duplicates()

print(product_info)
```

```python
import matplotlib.pyplot as plt

recommendations.plot(kind='bar', figsize=(8,5), color='purple')
plt.title("Top Recommended Products for " + target_customer)
plt.xlabel("Product ID")
plt.ylabel("Recommendation Score")
plt.show()
```

```python
def personalized_recommendation(target_customer, df, interaction_matrix, similarity_df, top_n=5):
    recs = recommend_products(target_customer, interaction_matrix, similarity_df, top_n)
    product_info = df[df['product_id'].isin(recs.index)][['product_id','product_name']].drop_duplicates()
    return product_info
```

```python
def hybrid_recommendation(target_customer, df, interaction_matrix, similarity_df, item_similarity_df, top_n=5):
    # Check if user has any purchases
    bought_products = interaction_matrix.loc[target_customer]
    
    if bought_products.sum() == 0:
        # Cold start → recommend popular products
        popular_products = df.groupby('product_id')['quantity'].sum().sort_values(ascending=False).head(top_n)
        product_info = df[df['product_id'].isin(popular_products.index)][['product_id','product_name']].drop_duplicates()
        product_info = product_info.set_index('product_id').join(popular_products.rename('score'))
        return product_info.sort_values('score', ascending=False)
    
    # Otherwise → blend user-based + item-based
    user_recs = recommend_products(target_customer, interaction_matrix, similarity_df, top_n*2)
    
    item_recs = pd.Series(dtype=float)
    for prod in bought_products[bought_products > 0].index:
        item_recs = item_recs.add(item_similarity_df[prod], fill_value=0)
    
    item_recs = item_recs[interaction_matrix.loc[target_customer] == 0]
    
    combined = user_recs.add(item_recs, fill_value=0).sort_values(ascending=False).head(top_n)
    
    product_info = df[df['product_id'].isin(combined.index)][['product_id','product_name']].drop_duplicates()
    product_info = product_info.set_index('product_id').join(combined.rename('score'))
    
    return product_info.sort_values('score', ascending=False)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

# Compute product similarity
item_similarity = cosine_similarity(interaction_matrix.T)

# Convert to DataFrame for readability
item_similarity_df = pd.DataFrame(item_similarity, 
                                  index=interaction_matrix.columns, 
                                  columns=interaction_matrix.columns)

# Example: check products similar to P1069 (Cookware Set)
print(item_similarity_df['P1069'].sort_values(ascending=False).head(6))
```

```python
import matplotlib.pyplot as plt

def plot_sales_trends(df):
    # Assuming df has columns: 'date' and 'quantity'
    sales_trend = df.groupby('date')['quantity'].sum()
    
    plt.figure(figsize=(10,5))
    sales_trend.plot(kind='line', marker='o', color='blue')
    plt.title("Sales Trend Over Time")
    plt.xlabel("Date")
    plt.ylabel("Total Sales")
    plt.grid(True)
    plt.show()
```

```python
def plot_category_distribution(df):
    # Assuming df has columns: 'category' and 'quantity'
    category_sales = df.groupby('category')['quantity'].sum()
    
    plt.figure(figsize=(7,7))
    category_sales.plot(kind='pie', autopct='%1.1f%%', startangle=90, colormap='tab20')
    plt.title("Sales Distribution by Category")
    plt.ylabel("")  # Hide y-label
    plt.show()
```

```python
def plot_segment_distribution(df):
    # Assuming df has columns: 'customer_segment' and 'customer_id'
    segment_counts = df.groupby('customer_segment')['customer_id'].nunique()
    
    plt.figure(figsize=(8,5))
    segment_counts.plot(kind='bar', color='green')
    plt.title("Customer Segmentation")
    plt.xlabel("Segment")
    plt.ylabel("Number of Customers")
    plt.xticks(rotation=45)
    plt.show()
```

```python
# Section 1: Sales Overview
plot_sales_trends(df)
plot_category_distribution(df)

# Section 2: Customer Segmentation
plot_segment_distribution(df)

# Section 3: Personalized Recommendations
recommendations_named = hybrid_recommendation('C000001', df, interaction_matrix, similarity_df, item_similarity_df, top_n=5)
recommendations_named['reason'] = ["Similar customers bought this" if score > 1 else "Popular fallback recommendation" for score in recommendations_named['score']]
display(recommendations_named[['product_name','score','reason']])
```

```python
import matplotlib.pyplot as plt

def full_dashboard(df, recommendations_named):
    fig, axes = plt.subplots(2, 2, figsize=(14,10))
    
    # --- Panel 1: Sales Trend ---
    sales_trend = df.groupby('transaction_date')['quantity'].sum()
    sales_trend.plot(kind='line', marker='o', color='blue', ax=axes[0,0])
    axes[0,0].set_title("Sales Trend Over Time")
    axes[0,0].set_xlabel("Transaction Date")
    axes[0,0].set_ylabel("Total Sales")
    axes[0,0].grid(True)

    # --- Panel 2: Category Distribution ---
    category_sales = df.groupby('category')['quantity'].sum()
    category_sales.plot(kind='pie', autopct='%1.1f%%', startangle=90, colormap='tab20', ax=axes[0,1])
    axes[0,1].set_title("Sales Distribution by Category")
    axes[0,1].set_ylabel("")  # Hide y-label

    # --- Panel 3: Customer Segmentation ---
    segment_counts = df.groupby('customer_segment')['customer_id'].nunique()
    segment_counts.plot(kind='bar', color='green', ax=axes[1,0])
    axes[1,0].set_title("Customer Segmentation")
    axes[1,0].set_xlabel("Segment")
    axes[1,0].set_ylabel("Number of Customers")
    axes[1,0].tick_params(axis='x', rotation=45)

    # --- Panel 4: Personalized Recommendations ---
    recommendations_named[['product_name','score']].set_index('product_name').plot(
        kind='bar', color='skyblue', ax=axes[1,1], legend=False
    )
    axes[1,1].set_title("Recommendations for Customer")
    axes[1,1].set_xlabel("Product Name")
    axes[1,1].set_ylabel("Score")
    axes[1,1].tick_params(axis='x', rotation=45)

    plt.tight_layout()
    plt.show()
```

```python
# Example: run full dashboard for C000001
recommendations_named = hybrid_recommendation('C000001', df, interaction_matrix, similarity_df, item_similarity_df, top_n=5)
recommendations_named['reason'] = ["Similar customers bought this" if score > 1 else "Popular fallback recommendation" for score in recommendations_named['score']]

full_dashboard(df, recommendations_named)
```
